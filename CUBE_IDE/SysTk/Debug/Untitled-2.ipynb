```python
hex_data = []       #the hex file is read here
Data = list()           #represents the 16 byte of data
Address = list()        #represents the low bytes of the address
CheckSum = list()

#function that reads the hex file and returns it as an array
def Read_Hex_File(Hex_path):
    with open(Hex_path, 'r') as f:
        return f.read()
    
def GET_Record_Type(Array, Index):
    if(Array[Index+7]+Array[Index+8] == '00'):
        return "data"
    elif (Array[Index+7]+Array[Index+8] == '01'):
        return "end"
    elif (Array[Index+7]+Array[Index+8] == '02'):
        return "obselete"
    elif (Array[Index+7]+Array[Index+8] == '03'):
        return "obselete"
    elif (Array[Index+7]+Array[Index+8] == '04'):
        return "high address"
    elif (Array[Index+7]+Array[Index+8] == '05'):
        return "full address"

def GET_Data_Size(Array, Index):
    return int((Array[Index+1] + Array[Index+2]), 16)

def GET_Address(Array, Index):
    sum = ''
    checksum = 0
    Address.append(Array[Index+3] + Array[Index+4] + Array[Index+5] + Array[Index+6])
    for i in range (3,7):
        sum = int((Array[Index+i]), 16)
        checksum +=sum
    return hex(checksum)

def GET_Data_Line(Array, Index, Data_size):
    line = ''
    sum = ''
    checksum = 0
    for i in range(9, 9 + Data_size*2):
        line += Array[Index+i] 
    for i in range(9, 9 + Data_size*2, 2):
        sum = int((Array[Index+i] + Array[Index+i+1]), 16)
        checksum +=sum
    Data.append(line)
    return hex(checksum)


def flip(c): 
    return '1' if (c == '0') else '0'
  
def printOneAndTwosComplement(bin): 
    n = len(bin) 
    line = '' 
    ones = "" 
    twos = "" 
    for i in range(n): 
        ones += flip(bin[i])  
    ones = list(ones.strip("")) 
    twos = list(ones) 
    for i in range(n - 1, -1, -1): 
      
        if (ones[i] == '1'): 
            twos[i] = '0'
        else:          
            twos[i] = '1'
            break
    i -= 1    
    if (i == -1): 
        twos.insert(0, '1')  
    for i in range (len(twos)):
        line += twos[i]
    return line
def Parse(Array):
    Addresschecksum = 0
    Datachecksum = 0
    Binary_sum = []
    checkSum = 0
    for i in range(len(Array)):
        if (Array[i] == ":"):
            if (GET_Record_Type(Array, i) == "data"):
                Datachecksum = GET_Data_Line(Array, i, GET_Data_Size(Array, i))
                Addresschecksum = GET_Address(Array, i)
                Binary_sum = bin(int(Datachecksum, 16) + int(Addresschecksum, 16))
                CheckSum.append(printOneAndTwosComplement(Binary_sum[2:]))





hex_data = Read_Hex_File('SysTk.hex')
#print(hex_data[16]+hex_data[17]+hex_data[18])
#print(GET_Data_Size(hex_data, 16))
#print(GET_Data_Line(hex_data, 16,16 ))
#print(Data)

Parse(hex_data)
for i in range(len(Data)):
    print("Data:  " + Data[i] +"  "+ "Address:  " + Address[i] +"  "+ "CheckSum:  " + CheckSum[i])
        
```
